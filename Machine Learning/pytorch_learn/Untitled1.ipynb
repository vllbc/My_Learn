```python
import numpy as np
import plotly.graph_objects as go
```

```python
trace1 = go.Scatter(
     y = np.random.randn(500),
    mode = 'markers',
    marker = dict(
        size = 16,
        color = np.random.randn(500),
        colorscale = 'Viridis',
        showscale = True
    )
)
fig = go.Figure(trace1)
fig.show()
```
