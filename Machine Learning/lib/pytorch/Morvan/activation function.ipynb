```python
import torch
import torch.nn
import torch.nn.functional as F
from torch.autograd import Variable
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
torch.__version__
```

```python
x = torch.linspace(-5,5,200)#从-5 到5随机200个点
x
```

```python
x = Variable(x)
```

```python
x_np = x.numpy()
#x_np = x.data.numpy()
```

```python
y_relu = F.relu(x).numpy()
plt.plot(x_np,y_relu)
plt.show()
```

```python
y_sigmoid = torch.sigmoid(x).numpy()
plt.plot(x_np,y_sigmoid)
plt.grid()
plt.show()
```

```python
y_tanh = torch.tanh(x).numpy()
plt.plot(x_np,y_tanh)
plt.show()
```

```python
y_softplus = F.softplus(x).numpy()
plt.plot(x_np,y_softplus)
plt.show()
```
