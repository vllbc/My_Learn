```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import json

import warnings
warnings.simplefilter("ignore")
```

```python
data1 = pd.read_csv("./data/估价训练数据.txt", sep="\t", header=None)
data2 = pd.read_csv("./data/门店交易训练数据.txt", sep="\t", header=None)
```

```python
data1_col = ["carid", "tradeTime", "brand", "serial", "model", "mileage", "color",
           "cityId", "carCode", "transferCount", "seatings", "registerDate", "licenseDate", "country",
           "maketype", "modelyear", "displacement", "gearbox", "oiltype", "newprice", ] + [f"anonymousFeature{n}" for n in range(1, 16)] + ["price"]
data2_col = ["carid", "pushDate", "pushPrice", "updatePriceTimeJson", "pullDate", "withdrawDate"]
data1.columns = data1_col
data2.columns = data2_col
```

```python
data = pd.merge(data1, data2, on="carid", how="inner") # 合并两个数据集
data.head()
```

```python
data["updatePriceTimeJson"] = data["updatePriceTimeJson"].apply(json.loads)
```

```python
data.info()
```

```python
data[["pushDate", "pullDate", "withdrawDate"]] = data[["pushDate", "pullDate", "withdrawDate"]].applymap(lambda x: pd.to_datetime(x))
```

```python
# 删除没有卖出去的数据
data.drop(data[data["withdrawDate"].isnull()].index, inplace=True)
```

```python
data["timedelta"] = (data["withdrawDate"] - data["pushDate"]).dt.days
data["timedelta"] = data["timedelta"].apply(lambda x: x if x > 0 else 0)
```

```python
data["timedelta"].describe()
```

可以看到成交周期最高为263天。平均为21天左右
